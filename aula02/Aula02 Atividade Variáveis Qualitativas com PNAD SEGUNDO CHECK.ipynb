```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_excel('PNAD 2013 A.xlsx')
```

```python
dados.head()
```

```python
n = dados.shape[0]
d = pd.DataFrame()
d["sexo"] = dados.V0302.astype('category')
d["cor_ou_raca"] = dados.V0404.astype('category')
d["rede_de_ensino"] = dados.V6002.astype('category')
d["nivel_de_ensino"] = dados.V4802.astype('category')
d["regiao"] = dados.UF
d["Estado_Civil"] = dados.V4011.astype('category')
d["Área_de_Ensino"] = dados.V6020.astype('category')
d["analfabeto"] = dados.V0601.astype('category')
```

```python
print('Frequências relativas por Sexo:')
d.sexo.cat.categories = ('Feminino','Masculino')
ut1 = ((d.sexo.value_counts()/n)*100).round(decimals=1)
print(ut1,'\n')
plot = ut1.plot(kind='pie',title='Sexo',autopct='%.1f',figsize=(6, 6))
```

```python
print('Frequências relativas por cor ou raça:')
d.cor_ou_raca.cat.categories = ('Indigena', 'Branca', 'Preta', 'Amarela', 'Pardo', 'Sem declaração')
ut2 = ((d.cor_ou_raca.value_counts()/n)*100).round(decimals=1)
print(ut2,'\n')
plot = ut2.plot(kind='bar',title='Raça ou Cor',figsize=(6, 6),color = ('b','w','k','r','y','c'))
plt.xlabel("cor ou raça")
plt.ylabel("frequência relativa em %")
```

```python
print('Frequências relativas por rede de ensino:')
d.rede_de_ensino.cat.categories = ('Pública', 'Privada')
ut3 = ((d.rede_de_ensino.value_counts()/n)*100).round(decimals=1)
print(ut3,'\n')
plot = ut3.plot(kind='bar',title='rede de ensino',figsize=(6, 6),color = ('b','w','k','r','y','c'))
plt.xlabel("rede de ensino")
plt.ylabel("frequência relativa em %")
```

```python
print('Frequências relativas por nivel_de_ensino:')
d.nivel_de_ensino.cat.categories = ("Maternal, jardim de infância etc.",
                              "Classe de alfabetização","Alfabetização de adultos",
                              "1ª a 4ª séries do ensino fundamental regular em 8 anos ou de educação de jovens e adultos"
                             ,"5ª a 8ª séries do ensino fundamental regular em 8 anos ou de educação de jovens e adultos"
                             ,"1ª série do ensino fundamental regular em 9 anos",
                              "2ª a 5ª séries do ensino fundamental regular em 9 anos",
                              "6ª a 9ª séries do ensino fundamental regular em 9 anos"
                             ,"Educação de jovens e adultos não seriado do ensino fundamental",
                              "Ensino médio, inclusive pré-vestibular","Superior, inclusive mestrado ou doutorado")
ut4 = ((d.nivel_de_ensino.value_counts()/n)*100).round(decimals=1)
print(ut4,'\n')
plot = ut4.plot(kind='bar',title='nivel de ensino',figsize=(6, 6),color = ('b','w','k','r','y','c'))
plt.xlabel("nivel_de_ensino")
plt.ylabel("frequência relativa em %")
```

```python
print('Frequências relativas por "Estado Civil":')
d.Estado_Civil.cat.categories = ('Solteiro','Casado', 'Separado Judicialmente', 'Divorciado', 'Viúvo')
ut6 = ((d.Estado_Civil.value_counts()/n)*100).round(decimals=1)
print(ut6,'\n')
plot = ut6.plot(kind='bar',title='Estado Civil',figsize=(6, 6),color = ('b','w','k','r','y','c'))
plt.xlabel("Estado Civil")
plt.ylabel("frequência relativa em %")
```

```python
print('Frequências relativas por Área de Ensino:')
d.Área_de_Ensino.cat.categories = ('Municipal', 'Estadual', 'Federal')
ut7 = ((d.Área_de_Ensino.value_counts()/n)*100).round(decimals=1)
print(ut7,'\n')
plot = ut7.plot(kind='bar',title='Área de Ensino',figsize=(6, 6),color = ('b','w','k','r','y','c'))
plt.xlabel("Área de Ensino")
plt.ylabel("frequência relativa em %")
```

```python
print('Frequências relativas por alfabetismo:')
d.analfabeto.cat.categories = ('Sim', 'Não')
ut99 = ((d.analfabeto.value_counts()/n)*100).round(decimals=1)
print(ut99,'\n')
plot = ut99.plot(kind='bar',title='Alfabetismo',figsize=(6, 6),color = ('b','w','k','r','y','c'))
plt.xlabel("Área de Ensino")
plt.ylabel("frequência relativa em %")
```

```python
ut20 = pd.crosstab(d.rede_de_ensino, d.cor_ou_raca)
print(ut20)
```

```python
#d["Regiao"] = dados.UF.astype('category')

d.regiao[(dados.UF >= 10) & (dados.UF < 20)] = 1
d.regiao[(dados.UF >= 20) & (dados.UF < 30)] = 2
d.regiao[(dados.UF >= 30) & (dados.UF < 40)] = 3
d.regiao[(dados.UF >= 40) & (dados.UF < 50)] = 4
d.regiao[(dados.UF >= 50) & (dados.UF < 60)] = 5

d.regiao = d.regiao.astype('category')
d.regiao.cat.categories = ('Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste')

ut25 = d.regiao.value_counts().round(decimals=1)
print(ut25,'\n')

plot = ut25.plot(kind='bar',title='Regiao',figsize=(6, 6),color=('b','g','k','r','m','c','y'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Regiões')
```

```python
ut24 = pd.crosstab(d.nivel_de_ensino, d.regiao)
plot = ut24.plot(kind='bar',title='Área de Ensino',figsize=(6, 6),color = ('b','w','k','r','y','c'))
plt.xlabel("Nível de ensino")
plt.ylabel("frequência relativa em %")
print(ut24)
```

```python
ut27 = pd.crosstab(d.analfabeto, d.regiao)
plot = ut27.plot(kind='bar',title='Área de Ensino',figsize=(6, 6),color = ('b','w','k','r','y','c'))
plt.xlabel("Sabe ler?")
plt.ylabel("frequência relativa em %")
print(ut27)
```
